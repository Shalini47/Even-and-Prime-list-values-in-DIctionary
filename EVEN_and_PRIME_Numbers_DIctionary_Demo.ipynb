```python
import numpy as np
```

```python
num=np.random.randint(1,5000,1000)
```

```python
num[0]
```

```python
Dictonary=dict()
even=list()
prime=list()
```

```python
for i in range(1000):
    if(num[i]%2==0):
        even.append(num[i])
    else:
        for j in range(2,num[i]):
            rem=num[i]%j
            if(rem==0):
                break
            if(j==num[i]-1):
                prime.append(num[i])           
```

```python
Dictonary["Even"]=even
Dictonary["Prime"]=prime
```
